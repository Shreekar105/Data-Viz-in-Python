## 1. Introduction on Cricket as a game 

•	  Cricket is a bat-and-ball game played between two teams of eleven players each on a field at the centre of which is a 22-yard (20-metre) pitch with a wicket at each end, each comprising two bails balanced on three stumps. 

•	  The game proceeds when a player on the fielding team, called the bowler, "bowls" (propels) the ball from one end of the pitch towards the wicket at the other end, with an "over" being completed once they have legally done so six times. 

•	The batting side has one player at each end of the pitch, with the player at the opposite end of the pitch from the bowler aiming to strike the ball with a bat. The batting side scores runs either when the ball reaches the boundary of the field, or when the two batters swap ends of the pitch, which results in one run. 

•	The fielding side's aim is to prevent run-scoring and dismiss each batter (so they are "out” and are said to have "lost their wicket"). Means of dismissal include being bowled, when the bowled ball hits the stumps and dislodges the bails, and by the fielding side either catching a hit ball before it touches the ground or hitting a wicket with the ball before a batter can cross the crease line in front of the wicket to complete a run. 

•	When ten batters have been dismissed, the innings ends and the teams swap roles. The game is adjudicated by two umpires, aided by a third umpire and match referee in international matches.


## 2. Introduction to IPL Auction 

•	The IPL auction is organized annually by the BCCI. Weeks before the IPL auction, a confirmed list of players who desire to play in the tournament is released. Given their talent and the characteristic that makes them stand out, the various franchises prepare and strategize to sign suitable players for their team. Apart from national players who represent their country, domestic cricketers are also scouted, which can give a huge break to a cricketer’s career.


•	Each team is assigned a budget, and within that budget, the franchise must bid one by one for the services of players from a particular category. Players are marked as Capped and Uncapped. A base price is set for each player who signs up for the IPL auction and makes it to the finalised list released by the BCCI. Each franchise owner, coach, and analyst analyzes and makes an offer for the player they want.


•	Only 161 players will be put under the hammer on the first day. The accelerated procedure will then begin on the second day, with the remaining 429 players being placed up for sale as well. Finally, the 10 teams will try to fill their squads with a maximum of 25 players while staying below the INR 90 crore budget.


•	The names of the players will be announced by the auctioneer, and the teams will begin bidding on them. Each cricketer will be auctioned off depending on their base price, with bidding continuing until the auctioneer declares the player sold. If no team bids for a player, they will go unsold. The auctioneer will give teams the option of listing the unsold players they are interested in, and the bidding for those players will be renewed with the base price lowered to half of what it was originally. The players will be considered unsold in the auction if they remain unclaimed for the second time.

## 4. Uploading / loading the Dataset

```python
# load data
data = pd.read_csv("ipl_2022_dataset.csv")
data.info()
data.head()
data.shape
data.columns
```

## 5.Cleaning the data 

```python
#Renaming all the column names as it was confusing 

data.columns = ['Index', 'Player','Base_Price','Type','CostInRupees(CR)','CostInDollars(DLR)',
              '2021_Squad','Team']
data.info()
data.head()
```

```python
#Removing unwanted column
data = data[['Player','Base_Price','Type','CostInRupees(CR)','CostInDollars(DLR)',
              '2021_Squad','Team']]
data.head()
```

```python
#Checking for Null values in the dataset 

data.isnull().sum()
```

## 6.Treating Null Values

```python
#Taking care NaN values in CostInRupees(CR)

data[data['CostInRupees(CR)'].isnull()]
```

```python
# Replacing null values with 0 for the players who went unsold in auction
```

```python
data['CostInRupees(CR)'] = data['CostInRupees(CR)'].fillna(0)
data['CostInDollars(DLR)'] = data['CostInDollars(DLR)'].fillna(0)
```

```python
# Taking care of NaN values in 2021_Squad Column
```

 Following were are the players who went unsold in 2021 IPL auction or are participating for the first time in IPL .

```python
data[data['2021_Squad'].isnull()]
```

```python
data['2021_Squad'] = data['2021_Squad'].fillna('Not Participated in IPL 2021')
```

```python
# Checking if any more null values exist
data.isnull().sum()
```

## 7.adding and Adjusting few columns for further Analysis

```python
teams = data[data['CostInRupees(CR)']>0]['Team'].unique()
teams
```

```python
data['status'] = data['Team'].replace(teams,'sold')
```

```python
data['Base_Price'].unique()
```

```python
data['retention'] = data['Base_Price']
```

```python
data['retention'].replace(['2 Cr', '40 Lakh', '20 Lakh', '1 Cr',
                         '75 Lakh','50 Lakh', '30 Lakh','1.5 Cr'],'In Auction', inplace = True)
```

Adjusting the Base Price Column

```python
data['Base_Price'].replace('Draft Pick',0, inplace = True)
```

```python
data['base_price_unit'] = data['Base_Price'].apply(lambda x: str(x).split(' ')[-1])
data['base price'] = data['Base_Price'].apply(lambda x: str(x).split(' ')[0])
```

```python
data['base price'].replace('Retained',0,inplace=True)
```

```python
data['base_price_unit'].unique()
```

```python
data['base_price_unit'] = data['base_price_unit'].replace({'Cr':100,'Lakh':1,'Retained':0})
```

```python
data['base price'] = data['base price'].astype(float)
data['base_price_unit'] = data['base_price_unit'].astype(int)
```

```python
data['base price'] = data['base price']*data['base_price_unit']
```

```python
data.tail()
```

```python
data.drop(['Base_Price','base_price_unit'], axis =1, inplace = True)
```

```python
data
```

```python
data['CostInRupees(CR)'] = data['CostInRupees(CR)']*100
```

```python
data = data.rename(columns={'TYPE':'Type','CostInRupees(CR)':'CostInRupees(Lakh)','2021_Squad':'Previous_team','Team':'Current_team'})
```
